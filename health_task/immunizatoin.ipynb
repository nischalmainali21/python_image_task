# Immunization Data Extraction

```python
import pandas as pd
import os
from tqdm import tqdm
import glob
```

```python
def extract_immunization_data(root_directory_path):
    file_paths = glob.glob(f"{root_directory_path}*.json")
    dfs = [pd.read_json(file_path) for file_path in file_paths]
    combined_df = pd.concat(dfs, ignore_index=True)
    entries = pd.json_normalize(combined_df['entry'])

    immunization_data = entries[entries['resource.resourceType'] == 'Immunization']
    immunization_data.reset_index(drop=True,inplace=True)

    immunization_status = immunization_data.get('resource.status',None)
    immunization_date = immunization_data.get('resource.date',None)
    immunization_name_list_raw = immunization_data.get('resource.vaccineCode.coding',None)
    if not immunization_name_list_raw is None:
        immunization_name_list_intermediate = pd.json_normalize(immunization_name_list_raw.explode())
        immunization_name_list = immunization_name_list_intermediate.get('display',None)
    immunization_patient_id = immunization_data['resource.patient.reference'].str.replace('urn:uuid:','')
    immunization_encounter_id = immunization_data['resource.encounter.reference'].str.replace('urn:uuid:','')
    intermediate_df = pd.DataFrame({
        'Status':immunization_status,
        'Immunization Date':immunization_date,
        'Immunization Name':immunization_name_list,	
        'Patient ID':immunization_patient_id,
        'Encounter ID':immunization_encounter_id
    })
    intermediate_df.to_csv('./output_csv/immunization_data.csv',index=False)
```

```python
extract_immunization_data('./health_data/healthcare-datasets/')
```
