## Procedure Data Extraction

```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
def extract_procedure_data(json_file_path):
    df = pd.read_json(json_file_path)
    entries = pd.json_normalize(df['entry'])

    # for patient id
    patient_data = entries[entries["resource.resourceType"]=="Patient"]
    patient_id = patient_data['resource.id']
    
    encounter_data = entries[entries["resource.resourceType"]=="Encounter"]

    procedure_data = entries[entries["resource.resourceType"]=="Procedure"]
    if not procedure_data.empty:
        procedure_data.reset_index(drop=True,inplace=True)
        procedure_name = procedure_data.get('resource.code.text',None)
        procedure_date_time = procedure_data.get('resource.performedDateTime',None)
        procedure_date_time_start = procedure_data.get('resource.performedPeriod.start',None)
        procedure_date_time_end = procedure_data.get('resource.performedPeriod.end',None)
        procedure_encounter_refrence = procedure_data['resource.encounter.reference']
        patient_id_series = pd.Series([patient_id[0]] * len(procedure_data))

        intermediate_df = pd.DataFrame({
            'Procedure Name': procedure_name,
            'Patient ID': patient_id_series,
            'Procedure Date Time': procedure_date_time,
            'Procedure Date Time Start':procedure_date_time_start,
            'Procedure Date Time End':procedure_date_time_end,
            'procedure_encounter_reference': procedure_encounter_refrence
        })

        encounter_conn_data = encounter_data[['resource.id','fullUrl']]

        merged_df = pd.merge(intermediate_df,encounter_conn_data,how='left',left_on='procedure_encounter_reference',right_on='fullUrl')
        merged_df.drop(columns=['procedure_encounter_reference','fullUrl'],inplace=True)
        merged_df.rename(columns={'resource.id':'Encounter ID'},inplace=True)
        merged_df = merged_df[['Procedure Name','Patient ID','Encounter ID','Procedure Date Time','Procedure Date Time Start','Procedure Date Time End']]

        return merged_df
    else:
        return None
```

```python
def procedure_data_csv(root_directory_path, output_csv_path):
    columns = ['Procedure Name','Patient ID','Encounter ID','Procedure Date Time','Procedure Date Time Start','Procedure Date Time End']
    with open(output_csv_path,'w') as f:
        pd.DataFrame(columns=columns).to_csv(f,index=False)
    for subdir, _, files in os.walk(root_directory_path):
        for file in tqdm(files):
            print(file)
            file_path = os.path.join(subdir, file)
            procedure_data = extract_procedure_data(file_path)
            if  procedure_data is not None:
                with open(output_csv_path, 'a') as f:
                    procedure_data.to_csv(
                        f, header=False, index=False)
```

```python
procedure_data_csv('./health_data/healthcare-datasets/','./output_csv/procedure_data.csv')
```
